```python
#Data pre-processing(Non numeric to numeric)

titanic['Sex'] = titanic['Sex'].apply(lambda x: 1 if x == 'male' else 0)
titanic['Sex']
```

```python
titanic
```

```python
#Missing values
titanic.isnull().sum()
```

```python
print(np.mean(titanic.Age))
#Fill with mean values
titanic['Age'] = titanic.Age.fillna(np.mean(titanic.Age))
```

```python
titanic['Age'].tail()
```

```python
titanic.Embarked.value_counts()
```

```python
#Filling with most number of values == 'S'
titanic.Embarked.fillna(value = 'S',axis= 0,inplace = True)
titanic.isnull().sum()
```

```python
titanic
```

```python
#Omit Irrelevant columns
titanic = titanic[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare']]
titanic
```

```python
#Seperate X and Y variables
X = titanic.drop('Survived', axis = 1)
Y = titanic['Survived']
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33, random_state =42)
```

```python
print(X_train.head())
print(X_test.head())
print(y_train.head())
print(y_test.head())
```

```python
#Building the model Decision tree

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
#Evaluate the model
!pip install graphviz
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,model.predict(X_train)))
print(accuracy_score(y_test,model.predict(X_test)))
```
